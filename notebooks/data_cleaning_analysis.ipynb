```python
import pandas as pd
df = pd.read_csv(r"F:\w\resume\strategic app\New folder\yellow_tripdata_2016-03.csv")
df.shape
```

```python
def filter_nyc_bounds(df):
    # NYC approximate boundaries
    nyc_min_lon, nyc_max_lon = -74.25, -73.70  
    nyc_min_lat, nyc_max_lat = 40.50, 41.1  

    
    
    constraint = (
        df['pickup_longitude'].between(nyc_min_lon, nyc_max_lon) &
        df['pickup_latitude'].between(nyc_min_lat, nyc_max_lat) &
        df['dropoff_longitude'].between(nyc_min_lon, nyc_max_lon) &
        df['dropoff_latitude'].between(nyc_min_lat, nyc_max_lat)
                  )
    return df[constraint]

df_clean = filter_nyc_bounds(df)
print(f"Removed {len(df) - len(df_clean)} outlier rows")
```

```python
# filtering trip_distance

df = df_clean[df_clean['trip_distance'].between(0.01,100)]
df.shape
```

```python
# dropping same duplicate trips

df = df.drop_duplicates(
    subset=['tpep_pickup_datetime', 'pickup_longitude', 'pickup_latitude'],
    keep='first' )
```

```python
df = df[df['passenger_count'].between(1, 6)]  
```

```python
# filling empty amounts 

df['fare_amount'] = df['fare_amount'].fillna(df['fare_amount'].median())
df['total_amount'] = df['total_amount'].fillna(df['total_amount'].median())
```

```python
# dropping unknown coordinates

df = df.dropna(subset=['pickup_longitude', 'pickup_latitude'])
df = df.dropna(subset=['dropoff_longitude', 'dropoff_latitude'])
```

```python
df.to_csv(r"F:\w\resume\strategic app\New folder\nyc_taxi_clean.csv")
print('Done!')
```
